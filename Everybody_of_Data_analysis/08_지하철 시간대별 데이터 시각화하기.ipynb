```python
# 출근시간 대 사람들이 가장 많이 타고 내리는 역은 어디?
# 출근시간 대를 오전 7시 

import csv 
f = open('subwaytime_202205.csv')
data = csv.reader(f)

next(data)
next(data)

result = []

for row in data: 
    sub = lambda x: int(x.replace(',',''))
    row[4:-1] = [sub(x) for x in row[4:-1]]
    result.append(row[10]) # 7시

print(len(result))
print()
print(result)
```

```python
# 출근시간 대 사람들이 가장 많이 타고 내리는 (max)역은 어디?
# 출근시간 대를 오전 7시 

import csv 
f = open('subwaytime_202205.csv')
data = csv.reader(f)

next(data)
next(data)

mx = 0
mx_station =''

for row in data: 
    sub = lambda x: int(x.replace(',',''))
    row[4:-1] = [sub(x) for x in row[4:-1]]
    
    if sum(row[10:15:2]) > mx:    # 승차를 가장 많이 하는 역
        mx = sum(row[10:15:2])
        mx_station = row[3] + '(' + row[1] + ')'

print(mx_station, mx)
```

```python
# 밤 11시에 사람들이 가장 많이 타는 역은 어디? 
# 입력값(특정 시간(t))받아 사람들이 가장 많이 타는 역 알아보기 

# 패턴 파악 : 시간(t)과 승차인원 index 관계 >> i = 4+ (t-4) * 2

import csv 
f = open('subwaytime_202205.csv')
data = csv.reader(f)

next(data)
next(data)

mx = 0
mx_station =""

t = int(input('몇 시 승차 인원이 가장 많은 역이 궁금하신가요?: '))


for row in data: 
    sub = lambda x: int(x.replace(',',''))
    row[4:-1] = [sub(x) for x in row[4:-1]]
    a = row[4+ (t-4) * 2]
    if a > mx:
        mx= a
        mx_station= row[3] + '(' + row[1] + ')'

print(mx_station, mx) # 승차인원이 가장 큰 역과 인원 입력 
```

```python
# 시간대별 사람들이 가장 많이 타고 내리는 역은 어디? 
# 하루는 24시간 (시간 단위 구분) >> range(24) 0 ~ 23 
# 시간단위 인덱스 t와 승차 인원 변수와의 관계식 >>> t*2+4 


import csv 
f = open('subwaytime_202205.csv')
data = csv.reader(f)

next(data)
next(data)

mx = [0] * 24           # 시간대별 최대 승차인원 저장 리스트 초기화
mx_station =['']*24     # 시간대별 최대 승차인원 역 이름 저장 리스트 초기화 

for row in data: 
    sub = lambda x: int(x.replace(',',''))
    row[4:-1] = [sub(x) for x in row[4:-1]]
    
    for t in range(24):   # 시간단위 구분(t) 0 ~ 23
        a = row[t*2+4]
        if a > mx[t]:
            mx[t] = a
            mx_station[t] = row[3] #+ '(' + row[1] + ')'

print(mx_station, mx)
```

```python
# 모든 지하철 역에서 시간대별 승하차 인원을 모두 더하면? 

import csv 
f = open('subwaytime_202205.csv')
data = csv.reader(f)

next(data)
next(data)

# 시간대별 승차인원을 저장할 리스트 초기화 
s_in = [0]*24

# 시간대별 하차인원을 저장할 리스트 초기화
s_out = [0]*24

for row in data: 
    sub = lambda x: int(x.replace(',',''))
    row[4:-1] = [sub(x) for x in row[4:-1]]
    
    for i in range(24):
        s_in[i] += row[4+i*2]
        s_out[i] += row[5+i*2]

plt.figure(figsize=(10,6), dpi=300)
plt.title('지하철 시간대별 승하차 인원 추이')
plt.plot(s_in, label='승차')
plt.plot(s_out, label='하차')
plt.legend()
plt.xticks(range(24), range(4,28))
plt.xlim([0,20])
plt.show()
```
