```python
# numpy 활용, 버블 차트 그리기 

x = np.random.randint(10,100,200)
y = np.random.randint(10,100,200)
# size = np.random.rand(1)*100
size = np.random.randint(10,100)

plt.scatter(x, y, s=size, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()

# randint(10,100,200)
# 10 ~ 100 정수(int) 200개가 random(무작위) 추출해서 생성
```

```python
import random

x = []
y = []
size = []

for i in range(200):
    x.append(random.randint(10,100))
    y.append(random.randint(10,100))
#     size.append(random.randint(10,100))
    size.append(random.randint(10, 100))
    
plt.scatter(x,y, s=size, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```

```python
### numpy 활용, 버블차트 그리기 

import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

```python
x  = np.random.randint(-100,100, 1000) # 1000개 무작위 추출 
y  = np.random.randint(-100,100, 1000)
size = np.random.rand(1)*100

mask1 = abs(x) > 50
mask2 = abs(y) > 50
x = x[mask1 + mask2] # mask1, mask2 하나라도 만족(or) 하는 값 저장 .
y = y[mask1 + mask2]

plt.scatter(x,y, s=size, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```
