```python
# 우리 동네 인구구조를 항아리 모양 그래프로 표현하기 

# gender(성별) 인구구조_항아리 모형 그리기 

import csv 
f = open("gender.csv")
data = csv.reader(f)

m = []
f = []

name = input('어디 사세요? 동네 이름을 알려주시면 인구 분포 알려줄게용~')

for row in data: 
    if name in row[0]: 
        for i in row[3:104]: 
            m.append(-int(i.replace(',',''))) # 겹쳐서 출력되기 때문에, (-) 부호 넣어서 음수로 변경
        for i in row[106:]:
            f.append(int(i.replace(',','')))
        break
        

print(m)
print(f)

import matplotlib.pyplot as plt
plt.title(name +'지역의 남녀 성별 인구분포')
plt.barh(range(101), m, label='남성')
plt.barh(range(101), f, label='여성')
plt.legend() # 범례
plt.show()
```

```python
# 우리 동네 인구구조를 항아리 모양 그래프로 표현하기 

# gender(성별) 인구구조_항아리 모형 그리기 

import csv 
f = open("gender.csv")
data = csv.reader(f)
 
size = []
# 남여 합계 구한 뒤 >> size 리스트에 성별 인구 합계 추가

name = input('어디 사세요? 동네 이름을 알려주시면 인구 분포 알려줄게용~')

for row in data: 
    if name in row[0]: 
        m = 0 
        f = 0
        
        for i in range(101):
            m += int(row[i+3].replace(',',''))
            f += int(row[i+106].replace(',',''))
        
        break
        

size.append(m)
size.append(f)
print(size)
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
color=['darkcyan', 'hotpink']
plt.axis('equal')
plt.pie(size, labels=['남','여'], autopct='%.1f%%', colors=color, 
       startangle=90)
# startangle : 반시계방향 90도로 이동
plt.title(name + '지역의 남녀 구성비율')
plt.show()
```

```python
# 꺽은선 그래프로 표현 

import csv 
f = open('gender.csv')
data = csv.reader(f)

m =[]
f =[]

name = input('남녀 인구분포가 궁금한 동네가 있나요?')
for row in data:
    if name in row[0]:
        for i in range(3, 104):
            m.append(int(row[i].replace(',','')))  # 남성 데이터 
            f.append(int(row[i+103].replace(',',''))) # 여성 데이터 
        break 
        
        
import matplotlib.pyplot as plt
plt.plot(m, label='male')
plt.plot(f, label='female')
plt.legend()
plt.show()
```

```python
# 막대 그래프로 표현 

import csv
f = open('gender.csv')
data = csv.reader(f)

result =[]
name = input('궁금한 동네를 입력해 주세요.')

for row in data: 
    if name in row[0]:
        for i in range(3,104):
            result.append(int(row[i].replace(',','')) - int(row[i+103].replace(',',''))) # 남성 - 여성
        break
```

```python
import matplotlib.pyplot as plt
plt.bar(range(101), result)
plt.show()
```

```python
# 결과 해석 : 있는 그대로 사실 
# 제주특별자치도 남녀 성비 분석 결과, 60대 기준(threshold), 
# 나이가 적은 인구는 남성이 많으나,60대 이상 인구는 여성이 확연히 많다 
# insight(통찰) >> 60대 여성의 복지 및 실버인구를 포용할 시설이 필요함
```

```python
# 연령대별 성별 비율을 산점도 표현하기 

import matplotlib.pyplot as plt
from astropy.visualization import astropy_mpl_style
astropy_mpl_style['axes.grid'] = False
plt.style.use(astropy_mpl_style) 
import math

import csv
f = open('gender.csv')
data = csv.reader(f)

m = []
f = []
size = []

name = input('궁금한 동네를 입력해 주세요.')

for row in data: 
    if name in row[0]:
        for i in range(3,104):
            m.append(int(row[i].replace(',','')))
            f.append(int(row[i+103].replace(',','')))
            size.append(math.sqrt(int(row[i].replace(',',''))+int(row[i+103].replace(',',''))))
        break


plt.style.use('ggplot')
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(10,6), dpi=300)
plt.title(name + '지역의 성별 인구 그래프')
plt.scatter(m,f, s=size, c=range(101), alpha=0.6, cmap='jet')
plt.colorbar()
plt.plot(range(max(m)), range(max(m)), 'blue')
plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')
plt.show()
```
