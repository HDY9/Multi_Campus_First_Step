```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# make the data
np.random.seed(3)
x = 4 + np.random.normal(0, 1, 24)  # 무작위 추출, 확률적 정규분포 
y = 4 + np.random.normal(0, 1, len(x))
# size and color:
sizes = np.random.uniform(15, 80, len(x)) # 무작위 추출, 균등 분할 
colors = np.random.uniform(15, 80, len(x)) 

# plot
fig, ax = plt.subplots()

ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
plt.title('legend') # legend : 범례
plt.plot([10,20,30,40], label='asc')
plt.plot([40,30,20,10], label='desc')
# plt.legend()
# plt.legend(loc = 'best')
plt.legend(loc='upper right')
plt.show()
```

```python
plt.title('color')
plt.plot([10,20,30,40], color='skyblue', label='skyblue') # 하늘색 그래프 그리기
plt.plot([40,30,20,10], color='pink', label='pink') # 핑크색 그래프 그리기 
plt.legend(loc='best')
plt.show()
```

```python
plt.title('line_style')
plt.plot([10,20,30,40], color='r', linestyle='--', label='dashed')
plt.plot([40,30,20,10], color='g', linestyle=':', label='dotted')
plt.legend()
plt.show()
```

```python
plt.title('marker')
plt.plot([10,20,30,40], 'r.', label='circle')
plt.plot([40,30,20,10], 'g^', label='triangle up')
plt.legend()
plt.show()
```
