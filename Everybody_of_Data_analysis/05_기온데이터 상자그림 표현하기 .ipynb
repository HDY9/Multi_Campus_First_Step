```python
import csv 
f = open('seoul.csv')
data = csv.reader(f)
next(data) # header 지나서 data로 접근

# print(data)

# result = []

aug =[]
jan =[]

for row in data: 
    month = row[0].split('-')[1]
    if row[-1] != '': 
        if month == '08':
            aug.append(float(row[-1]))
        if month == '01': 
            jan.append(float(row[-1]))

        
import matplotlib.pyplot as plt 
# plt.boxplot(result)
plt.boxplot(aug)
plt.boxplot(jan)
plt.show()
```

```python
import matplotlib.pyplot as plt 
import csv 

f = open('seoul.csv')
data = csv.reader(f)
next(data)
```

```python
# 월별 데이터를 저장할 리스트 month (12개)

month = [[],[],[],[],[],[],[],[],[],[],[],[]]
```

```python
for row in data: 
    if row[-1]!='': 
        month[int(row[0].split('-')[1])-1].append(float(row[-1]))
        # 1월 ~ 12월 >> month[0] : 1월 
        
plt.boxplot(month)
plt.show()    
```

#### 8월 / 1월 일별 기온데이터를 box plot으로 표현하기 



```python
import matplotlib.pyplot as plt 
import csv
```

```python
f = open('seoul.csv')
data = csv.reader(f)
next(data)

day = [] # 일별 데이터를 저장할 list 

for i in range(31): 
    day.append([])
    
for row in data:
    if row[-1] !='': 
        if row[0].split('-')[1] == '08': 
            #최고 기온 값 저장 
            day[int(row[0].split('-')[2])-1].append(float(row[-1]))
            
plt.style.use('ggplot') 
plt.figure(figsize=(10,5), dpi=300) 
# plt.boxplot(day, showfliers=False) # outliers(이상치) 생략 
plt.boxplot(day)
plt.show()
```
