```python
import csv 

f = open('seoul.csv', 'r', encoding='cp949')
data = csv.reader(f)

next(data)
result = [ ]

for row in data: 
#     print(row)
#     print(row[-1])  
    if row[-1]!='': #  최고 기온이 빈 문자열이 아니면(데이터 값이 존재하면)
        result.append(float(row[-1]))

print(result)
print()


# data로만 접근(next()),최고 기온만 추출  
```

```python
import csv 

f = open('seoul.csv', 'r', encoding='cp949')
data = csv.reader(f)

next(data)
high = [ ] # 최고 기온 변화를 담을 리스트
low = [ ]  # 최저 기온 변화를 담을 리스트 

for row in data: 
#     print(row)
#     print(row[-1])  
    if row[-1]!='' and row[-2]!='':  # 최저기온, 최고기온이 존재한다면, 
        if 2000 <= int(row[0].split('-')[0]):
            if row[0].split('-')[1] == '10' and row[0].split('-')[2] == '26':
                high.append(float(row[-1]))
                low.append(float(row[-2]))
        
plt.plot(high, 'red')        
plt.plot(low, 'blue')
plt.rc('font', family='Malgun Gothic')
plt.title('내 생일의 기온변화 그래프')
plt.show()
```

```python
import csv 

f = open('seoul.csv', 'r', encoding='cp949')
data = csv.reader(f)

next(data)
high = [ ] # 최고 기온 변화를 담을 리스트
low = [ ]  # 최저 기온 변화를 담을 리스트 


for row in data: 
#     print(row)
#     print(row[-1])  
    if row[-1]!='' and row[-2]!='':  # 최저기온, 최고기온이 데이터가 존재한다면, 
        if 2000 <= int(row[0].split('-')[0]): # 2000년 이후 데이터라면, 
            if row[0].split('-')[1] == '10' and row[0].split('-')[2] == '26':
                high.append(float(row[-1]))
                low.append(float(row[-2]))
        
plt.plot(high, 'red', label='high')        
plt.plot(low, 'blue', label='low')

plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
plt.rc('font', family='Malgun Gothic')
plt.title('내 생일의 기온변화 그래프')
plt.legend() # 범례
plt.show()

f.close()
```
