```python
pd.read_excel('sample_codemaster.xlsx')
```

```python
df_codemaster = pd.read_excel('sample_codemaster.xlsx')
```

```python
# inner join

pd.merge(left = df, 
         right = df_codemaster,
#          how ='left', # default : inner join (교집합) cf. outer join(합집합)
         left_on = '국적코드',
         right_on ='국적코드'
        )
```

```python
# outer join

pd.merge(left = df, 
         right = df_codemaster,
         how ='outer', # default : inner join (교집합) cf. outer join(합집합)
         left_on = '국적코드',
         right_on ='국적코드'
        )
```

```python
pd.merge(left = df, 
         right = df_codemaster,
         how ='left',
         left_on = '국적코드',
         right_on ='국적코드'
        )
```

```python
pd.merge(left = df, 
         right = df_codemaster,
         how ='right',
         left_on = '국적코드',
         right_on ='국적코드'
        )
```
