# 삼겹살 가격 분석

```python
df[(df['품목 이름'].str.contains('삼겹살')) & (df['년도-월']=='2021-06')] 
```

```python
# 600g 만 

df[(df['품목 이름'].str.contains('삼겹살')) & (df['년도-월']=='2021-06') &
  (df['실판매규격'].str.contains('600g'))] 
```

```python
df_sam = df[(df['품목 이름'].str.contains('삼겹살')) & (df['년도-월']=='2021-06') &
  (df['실판매규격'].str.contains('600g'))] 
```

```python
df_sam.head(2)
```

```python
# 삼겹살 600g의 평균 가격은?
df_sam['가격(원)'].mean()
```

```python
# 삼겹살 600g의 최고 가격은?
df_sam['가격(원)'].max()
```

```python
df_sam[df_sam['가격(원)'] == 35890]
```

```python
# 삼겹살 600g의 최저 가격은?
df_sam['가격(원)'].min()
```

## 우리동네 삼겹살 가격

```python
gu =input('구 이름: ')
```

```python
df_sam.head(1)
```

```python
# 우리구 삼겹살 가격
df_sam[df_sam['자치구 이름'] == gu][['시장/마트 이름','품목 이름', '실판매규격','가격(원)']].drop_duplicates()
```

```python
df_sam_gu = \
df_sam[df_sam['자치구 이름'] == gu][['시장/마트 이름','품목 이름', '실판매규격','가격(원)']].drop_duplicates()
```

```python
# 시각화
x = df_sam_gu['시장/마트 이름']
y = df_sam_gu['가격(원)']
plt.scatter(x, y)
plt.title(gu + '삼겹살 가격 현황')
plt.grid()
plt.show()
```
