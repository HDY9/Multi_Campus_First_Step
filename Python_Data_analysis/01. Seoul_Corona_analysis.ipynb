# 구 별 확진자 동향

## 확진일-구별 확진자수 집계

### 피봇테이블 만들기

```python
pd.pivot_table(df,index='확진일', columns='지역', values='연번', aggfunc='count', margins=True)
# '연번' count >>> 확진자수 세기, margins=True : 각 행/열 대한 합계를 가져오기 
```

```python
df_gu = pd.pivot_table(df,index='확진일', columns='지역', values='연번', aggfunc='count', margins=True)
```

### 서울시 일별 추가확진자 동향

```python
df_gu['All']
```

```python
df_gu['All'][:-1]
```

```python
s_date = df_gu['All'][:-1]
```

```python
# 서울시 일별 추가확진자가 많았던 순으로 보기
s_date.sort_values(ascending=False)
# ascending=False 내림차순
```

```python
s_date.index
```

```python
# s_date.values
```

```python
# 서울시 일별 추가확진자 시각화

x = s_date.index
y = s_date.values

plt.plot(x,y)
plt.title("서울시 일별 추가확진자 현황(2021.09.28 까지)")
plt.xlabel("확진일")
plt.ylabel("추가확진자 수")
plt.xticks(rotation=45)
plt.show()
```

### 서울시 구별 누적확진자 비교

```python
# 서울시 구별 누적확진자 많은 순으로 보기
df_gu
```

```python
df_gu.loc['All']
```

```python
df_gu.loc['All'][:-1]
```

```python
s_gu = df_gu.loc['All'][:-1]
s_gu.sort_values(ascending=False)
```

```python
s_gu = s_gu.sort_values(ascending=False)
s_gu
```

```python
# 서울시 구별 누적확진자 많은 순으로 시각화
```

```python
x = s_gu.index
y = s_gu.values

plt.figure(figsize = (12,7))
plt.title("서울시 구별 누적확진자(2021.9.28 까지)", size = 20)
plt.barh(x,y)
plt.show()
```

### 최근일 기준 지역별 추가확진자 

```python
df_gu
```

```python
df_gu.iloc[-2][:-1]
```
