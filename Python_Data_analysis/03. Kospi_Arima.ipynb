```python
series = pd.read_csv('../data/kospi3.csv', header=0, index_col=0)
series.plot()
```

```python
import matplotlib.pyplot as plt 
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
# acf, pacf 곡선 그리기 
plot_acf(series)
plot_pacf(series)
plt.show()

# 결과 해석 
# ACF의 첫 번째 구간 : 영향력을 미치는 구간
# ACF 의 실내 공간 : 영향력이 없음 

# PACF 첫 번째 영향구간
```

```python
series.head()
```

```python
# 차분곡선 그래프 (1차 차분: 1번 차분)
 
series.diff(periods=1).iloc[1:]
```

```python
series.plot()
```

```python
diff_1 = series.diff(periods=1).iloc[1:]
diff_1.plot()
```

```python
plot_acf(diff_1) # ma
plot_pacf(diff_1) # ar
plt.show()
```

```python
# 결과 해석 

# 평균값이 이동하는 모습을 보임(비정상)
# ACF : 하나만(1 : 자기자신), 나머지는 실내공간에 포진됨 
# PACF : 대부분 실내공간에 있다가, 들쭉날쭉한 모습이 관찰됨
```

```python
# 차분의 차분 그래프 그리기 

diff_1.diff(periods=1).iloc[1:]
```

```python
diff_2 = diff_1.diff(periods=1).iloc[1:]
diff_2.plot()
# 평균값이 거의 같은 모습을 띔 (안정화된 모형 >> 정상성 )
```

```python
plot_acf(diff_2)
plot_pacf(diff_2)
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA

# ARIMA(1,2,1) 모델 생성하고 훈련하기 

model = ARIMA(series, order=(1,2,1)) # (p,d,q) AR + I + MA : 차분을 두 번: d를 2로 설정  
model_fit=model.fit()
print(model_fit.summary())

# 결과 해석 
# P>|z| (p-value): ar 0.951 값이 좋지 않음, ma 0.00
# ar 삭제 >> 0 (0,2,1)
```

```python
model = ARIMA(series, order=(0,2,1)) # (p,d,q) AR + I + MA : 차분을 두 번: d를 2로 설정  
model_fit=model.fit()
print(model_fit.summary())
```

```python
model_fit.plot_diagnostics()
```

```python
model_fit.predict().plot()
```

```python
model_fit.forecast(steps=1) # 1단계 시점(t+1) 예측
```

```python
model_fit.forecast(steps=10)
```
