## 1. 1차원 정렬

1차원 정렬은 매우 간단합니다.

```
import numpy as np
arr = np.array([1, 10, 5, 8, 2, 4, 3, 6, 8, 7, 9])
```

```
arr
```

np.sort(arr)

```
np.sort(arr)
```

```
# 순정렬 
np.sort(arr)[::1]
```

기본적으로 오름차순 정렬을 수행합니다.

```
np.sort(arr)[::-1]
```

내림차순 정렬은 옵션에 -1 값을 주면됩니다.

하지만, 이 값들은 **유지가 되지 않습니다**!

값을 sort 된 상태로 저장하려면, 배열 자체에 sort를 해줄 수 있습니다.

```
np.sort(arr)
```

```
arr2 = np.sort(arr)
```

## 2. N차원 정렬

```
arr2d = np.array([[5, 6, 7, 8], 
                  [4, 3, 2, 1],
                  [10, 9, 12, 11]])
```

```
arr2d.shape
```

```
arr2d
```

### 2-1. 열 정렬 (왼쪽에서 오른쪽으로)

```
arr2d
```

```
np.sort(arr2d, axis=1) # 열기준(서로 다른 열기리), 가로방향
```

### 2-2. 행 정렬 (위에서 아래로)

```
arr2d
```

```
np.sort(arr2d, axis=0)
```

## 3. index를 반환하는 argsort

정렬된 값을 반환하는 것이 아닌 index를 반환합니다.

```
arr2d = np.array([[5, 6, 7, 8], 
                  [4, 3, 2, 1],
                  [10, 9, 12, 11]])
```


### 3-1. 열 정렬 (왼쪽에서 오른쪽으로)

```
np.argsort(arr2d, axis=1)  # 서로 다른 열끼리, 같은 행끼리 비교 (가로방향)
```

### 3-2. 행 정렬

```
arr2d
```

```
np.sort(arr2d, axis=0)  # 서로 다른 행끼리, 같은 열끼리 비교 (세로방향)
```

```
np.argsort(arr2d, axis=0)
```

```
name = np.array(['쯔위','수지','카즈하','진욱'])
score = np.array([95,97,100,99])
np.argsort(name) 
```

```
np.argsort(score)
```

```
score_index = np.argsort(score)[::-1] # 내림차순 정렬
score_index
```

```
name[score_index]  # 성적 순 정렬 
```
