```
import numpy as np
```

## 1. 행렬 - 덧셈

### 1-1. 덧셈

```
a = np.array([[1, 2, 3], 
              [2, 3, 4]])
```

```
b = np.array([[3, 4, 5], 
              [1, 2, 3]])
```

```
a + b
```

```
a = np.array([[1, 2, 3], 
              [2, 3, 4]])
```

```
b = np.array([[1, 2],
              [3, 4],  
              [5, 6]])
```

```
# a + b
# ValueError 발생
```

### 1-2. Sum

```
a = np.array([[1, 2, 3], 
              [2, 3, 4]])
```

```
np.sum(a, axis=0) # 행 끼리, 서로 다른 행끼리(세로방향)
```

```
np.sum(a, axis=1) # 열 끼리, 서로 다른 열끼리(가로방향)
```

## 행렬곱연산

```
a = np.array([[1, 2, 3],
              [1, 2, 3], 
              [2, 3, 4]])
```

```
b = np.array([[1, 2],
              [3, 4],  
              [5, 6]])
```

```
a.shape, b.shape
```

```
# 행렬곱을 할때 dot메소드 사용
np.dot(a, b)
```

```
a.dot(b)
```
