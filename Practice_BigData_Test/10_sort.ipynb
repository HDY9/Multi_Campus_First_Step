```python
emp = pd.read_csv('emp.csv')

emp_copy = emp.copy()
```

### 정렬 
1. sort_index 
- Series, DataFrame 호출 가능 
- index, column 재배치 

```python
emp = emp.set_index('empno')
emp
```

```python
emp.sort_index(ascending=False)
```

```python
emp.sort_index(axis=1)
```

2. sort_values 
- Series, DataFrame 호출 가능 
- 본문의 값으로 정렬 가능(Series 값 또는 DataFrame 특징 컬럼 순으로)

```python
emp
```

```python
emp.sort_values(by='sal') # 오름차순
```

```python
emp.sort_values(by='sal',ascending=False) # 내림차순
```

```python
emp.sort_values(by=['deptno','sal'], ascending=False)
```

```python
# deptno 올림차순, sal 내림차순
emp.sort_values(by=['deptno','sal'], ascending=[True, False])
```

```python
emp.sort_values(by=['deptno','sal'], ascending= [False, True])
```
