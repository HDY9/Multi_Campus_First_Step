```python
# -*- coding: UTF-8 -*-
import time
from selenium import webdriver
 
#Colab에선 웹브라우저 창이 뜨지 않으므로 별도 설정한다.

# webdriver 기본 option 생성
options = webdriver.ChromeOptions()
options.add_argument('--headless')        
# Head-less 설정(백그라운드에 돌아가게 하기 위함)
options.add_argument('--no-sandbox')  
# 리눅스 서버와 같이 GUI 제공하지 않는 환경에서 추가해야 함 
options.add_argument('--disable-dev-shm-usage')
# 보안 관련 옵션
driver = webdriver.Chrome('chromedriver', options=options)
 
#해당 url로 이동
url = "https://www.naver.com/" 
driver.get(url)
 
update = driver.find_element_by_css_selector('#NM_TS_ROLLING_WRAP > div > div')
print(update.text)
# NM_TS_ROLLING_WRAP
```

```python
# 예제 2-63 멜론 인기차트 중 상위 50곡 크롤링(정리)  # 멜론 사이트 개편으로 코드 수정(updated 2020.07.13)
  
url = 'http://www.melon.com/chart/index.htm'
driver.get(url)          
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
                               
songs = soup.select('tr')[1:]     # 멜론 사이트 개편으로 코드 수정(updated 2020.07.13)
for song in songs:                                        
    title = song.select('div.ellipsis.rank01 > span > a')[0].text
    singer = song.select ('div.ellipsis.rank02 > a')[0].text  
    print(title, singer, sep = ' , ')
```

```python
# csv 파일로 저장 

url = 'http://www.melon.com/chart/index.htm'
driver.get(url)          
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

song_title = []
song_singer = []   
song_contents =[]

songs = soup.select('tr')[1:]     # 멜론 사이트 개편으로 코드 수정(updated 2020.07.13)
for song in songs:                                        
    title = song.select('div.ellipsis.rank01 > span > a')[0].text
    singer = song.select ('div.ellipsis.rank02 > a')[0].text  
    song_title.append(title)
    song_singer.append(singer)
    song_contents.append(title + ','+singer)
    print(song_contents)
```

### 2.2.13 selenium을 활용한 크롤링

```python
# 예제 2-64 멜론 인기차트 중 상위 50곡 크롤링(selenium만 활용)
     
url = 'http://www.melon.com/chart/index.htm'
driver.get(url)          
                                
songs = driver.find_elements_by_css_selector('tr')[1:]     # 멜론 사이트 개편으로 코드 수정(updated 2020.07.13)
for song in songs:                                        
    title = song.find_elements_by_css_selector('div.ellipsis.rank01 > span > a')[0].text
    singer = song.find_elements_by_css_selector('div.ellipsis.rank02 > a')[0].text  
    print(title, singer, sep = ' | ')
```
